```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# Load the data
df = pd.read_csv("data-1777933083910.csv")

# Select only the columns you want to impute
columns_to_impute = ['i1_speed', 'i2_speed', 'st_speed']
df_subset = df[columns_to_impute]

# Scale the data (important for KNN)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_subset)

# Apply KNN imputation
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform')
df_imputed_scaled = knn_imputer.fit_transform(df_scaled)

# Transform back to original scale
df_imputed = scaler.inverse_transform(df_imputed_scaled)

# Replace the original columns with imputed values
df[columns_to_impute] = df_imputed

# CRITICAL: Ensure data types match your PostgreSQL schema
df['id'] = df['id'].astype('Int64')  # Nullable integer
df['meeting_key'] = df['meeting_key'].astype('Int64')
df['session_key'] = df['session_key'].astype('Int64')
df['driver_number'] = df['driver_number'].astype('Int64')
df['lap_number'] = df['lap_number'].astype('Int64')
df['i1_speed'] = df['i1_speed'].astype('float64').round(1)
df['i2_speed'] = df['i2_speed'].astype('float64').round(1)
df['st_speed'] = df['st_speed'].astype('float64').round(1)
df['duration_sector_1'] = df['duration_sector_1'].astype('float64').round(3)
df['duration_sector_2'] = df['duration_sector_2'].astype('float64').round(3)
df['duration_sector_3'] = df['duration_sector_3'].astype('float64').round(3)
df['lap_duration'] = df['lap_duration'].astype('float64').round(3)
df['is_pit_out_lap'] = df['is_pit_out_lap'].astype('boolean')

# Save the COMPLETE clean file
df.to_csv('laps_clean_for_pentaho.csv', index=False)

print("✅ Saved clean file: laps_clean_for_pentaho.csv")
print(f"✅ Shape: {df.shape}")
print(f"✅ Nulls remaining: {df[columns_to_impute].isnull().sum().sum()}")
```
